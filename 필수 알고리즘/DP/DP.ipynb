
# DP 알고리즘
복잡한 문제를 여러 개의 문제를 나눠서 푸는 기법

메모이제이션 기법을 활용해서 중복된 계산을 줄이는 것으로 답을 구한다.

문제의 재귀적인 구조(관계식)을 찾아야 한다.

## STEP
1. 문제의 재귀적인 구조 파악
2. 재귀적인 구조를 수식으로 표현
3. 초기값 처리

## 주의사항
DP 알고리즘은 적용해야할 상황이 정해져 있지 않기 때문에 문제를 풀어보면서 감을 익혀야 한다.

만약, 재귀적인 구조를 찾지 못했다면? 문제를 재해석하여 재귀 구조를 파악하거나 DP 가 아닌 브루트 포스나 그리디, 수학적인 풀이를 생각해야한다.

## 풀이 팁
점화식을 만들 때, 답에서 부터 하나씩 들어가는 발상을 해봐야 한다.
어떻게 하면 답이 만들어지는 지에 대해서 생각해보자.

## 예시
동전 1, 3, 4 가 있을 때, 각 동전을 최소로 사용하여 목표 금액을 맞출 수 있을 지 알아보는 문제를 푼다고 해보자.

(아마 평소라면 1, 3, 4를 전부 고려해서 사용하는 방식으로 문제를 풀려고 했을 것이다.)

그런데 만약에 목표 금액에 맞춰서 점화식을 작성한다면?

`목표 금액 i = min(f(i -1), f(i - 3), f(i - 4)) + 1` 이라는 점화식이 생성된다. (조금 고민해보면 이 점화식의 의미를 파악할 수 있다.)

이 때 중요한 것은 관계와 예외처리 이다. 함수 f의 매개변수가 음수인 것만 처리해주면 된다.

그리고 이대로면 재귀구조이니까 연산량이 많아진다. 이 때, 연산량을 줄이기 위해서 배열로 연산의 결과를 저장하고 있으면 빠르게 연산할 수 있다.
